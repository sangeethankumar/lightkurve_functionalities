## TIC 140757590

```python
lc = pprop.searchNstitch(140757590)
ls,bls = pprop.genp(lc)
```

```python
lc.plot()
```

```python
ax = ls.plot(scale='log')
```

```python
ls_period,bls_period,t0,dur = pprop.getLSparams(ls,bls)
print(ls_period,bls_period,t0,dur)
```

```python
lc.fold(period=ls_period).plot()
lc.fold(period=bls_period,epoch_time=t0).plot()
```

```python
lc.fold(period=ls_period,epoch_time=t0).plot_river()
```

```python
lc.fold(period=bls_period,epoch_time=t0).plot_river()
```
